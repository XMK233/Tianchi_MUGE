```python
import os
import torch
from PIL import Image
import torchvision.transforms as T
from omegaconf import OmegaConf
from taming.models.vqgan import VQModel

def load_vqgan_model(config_path, checkpoint_path):
    """加载预训练 VQGAN 模型"""
    config = OmegaConf.load(config_path)
    model = VQModel(**config.model.params)
    state_dict = torch.load(checkpoint_path, map_location="cpu")["state_dict"]
    model.load_state_dict(state_dict, strict=True)
    model.eval()
    return model

# === 配置路径 ===
CONFIG_PATH = "/mnt/d/HuggingFaceModels/vqgan_imagenet_f16_1024/config.yaml"
CKPT_PATH = "/mnt/d/HuggingFaceModels/vqgan_imagenet_f16_1024/model.ckpt"

# === 加载模型 ===
device = "cuda" if torch.cuda.is_available() else "cpu"
vqgan = load_vqgan_model(CONFIG_PATH, CKPT_PATH).to(device)

# === 图像预处理 ===
transform = T.Compose([
    T.Resize((256, 256)),
    T.ToTensor(),
    T.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5]),  # [-1, 1]
])

# === 示例：加载一张图片并编码/解码 ===
img_path = "example.jpg"  # 替换为你的商品图路径
img_pil = Image.open(img_path).convert("RGB")
img_tensor = transform(img_pil).unsqueeze(0).to(device)  # [1, 3, 256, 256]

# 编码：图像 → 离散 token (indices)
with torch.no_grad():
    z, _, [_, _, indices] = vqgan.encode(img_tensor)
    print("Token shape:", indices.shape)  # 应为 [256]（16x16）
    
    # 解码：token → 图像
    quant = vqgan.quantize.embedding(indices).view(1, 16, 16, -1).permute(0, 3, 1, 2)
    rec_img = vqgan.decode(quant)

# 后处理重建图像
rec_img = torch.clamp(rec_img, -1.0, 1.0)
rec_img = (rec_img + 1.0) / 2.0  # [0, 1]
rec_pil = T.ToPILImage()(rec_img.squeeze(0).cpu())

# 保存重建图
rec_pil.save("reconstructed.png")
print("原始图 -> 编码 -> 解码 重建完成！")
```
